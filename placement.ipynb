```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('modified_placement_data.csv')
```

```python
df.info()
```

```python
updated_df=df.fillna(df['cgpa'].median)
```

```python
updated_df
```

```python
updated_df.info()
```

```python
X = updated_df.iloc[:,:2]
Y = updated_df['placement']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, Y, test_size= 0.2)
```

```python
X_train['cgpa'] = pd.to_numeric(X_train['cgpa'], errors='coerce')
```

```python
X_train['cgpa'].fillna(X_train['cgpa'].mean(), inplace=True)  # Replace NaN with mean
```

```python
X_train['cgpa'] = X_train['cgpa'].astype(float)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test
```

```python
X_test['cgpa'] = pd.to_numeric(X_test['cgpa'], errors='coerce')
```

```python
X_test['cgpa'].fillna(X_test['cgpa'].mean(), inplace=True)  # Replace NaN with mean
```

```python
X_test['cgpa'] = X_test['cgpa'].astype(float)
```

```python
scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
